#### 1. Limpieza 

Limpiar los documentos de caracteres que no corresponden

```python
doc = df.iloc[0]['review']
doc
```

```python
doc = doc.replace('<br />', '').replace('>', '').replace('/', '')
```

```python
import re
```

```python
def clean_text(doc):
    return re.sub(pattern=r'<.*?>', repl='', string=doc).replace('.', ' ').replace(',', '').replace('(', '').replace(')', '').replace('"', '').replace("'", '').replace("\x08", '')
```

```python
df['review'].apply(clean_text)
```

```python
doc.split()
```

#### 2. Normalización

Convertir todos los tokens a minúsculas.

Elimina puntuación y símbolos no alfabéticos.

```python
# Convierto todo a minisculas y elimina lo que no sea simbolos alfanumericos
def normalized(texto):
    texto = texto.lower()
    texto_limpio = re.sub(r'[^a-z0-9\s]', '', texto)
    return texto_limpio
```

```python
# pruebo a ver si funciona con un documento
doc = normalized(doc)
doc
```

```python
#aplico a todo el corpus la normalizaacion
df['review'].apply(normalized)
```

#### 3. Eliminación de Stopwords

Eliminar las palabras vacías (stopwords) usando una lista estándar de la librería _nltk_.

* filtra y tokeniza las palabras mas importantes del corpus

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def filtrar_stopwords(text):
    # tokeniza las palabras de un documento
    tokes = word_tokenize(text)
    # obtiene los stopwords del ingles
    en_stop_words = set(stopwords.words('english'))
    # filtra los token con las stopwords
    return [t for t in tokes if t.isalpha() and t not in en_stop_words]
```

```python
filtrar_stopwords(doc)
```

#### 4. Stemming 

Reducir el espacio de palabras

```python
from nltk.stem import porter

stemmer = porter.PorterStemmer()
```

```python
def get_stemming(doc):
    # obtiene el documento con las stopwords eliminadas
    palabras = filtrar_stopwords(doc)
    # aplica stemming a las palabras
    return [stemmer.stem(p) for p in palabras]
```

```python
# imprime el documento preprocesado
print(' '.join(get_stemming(doc)))
```

#### 5. Verificar la diferencia

Comparar el tamaño del diccionario de términos del corpus antes y después de aplicar el preprocesamiento 

```python
dicc_original = set()
# Crea el diccionario con el corpus original
for docu in df['review']:
    for palabra in docu.split():
        dicc_original.add(palabra)
```

```python
len(dicc_original)
```

```python
stop = set(stopwords.words('english'))
stemmer = porter.PorterStemmer()
# funcion para preprocesar con las funciones creadas
def preprocesar(texto):
    texto = clean_text(texto)
    texto = normalized(texto)
    tokens = get_stemming(texto)
    return tokens
```

```python
dicc_preprocesado = set()
# Crea el diccionario con los documentos preprocesados
for docu in df['review']:
    tokens = preprocesar(docu)
    dicc_preprocesado.update(tokens)
```
