## Leer el documento txt

```python
path = 'data/01_corpus_turismo_500.txt'
with open(path,'r', encoding='utf-8') as f:
    text = f.read()
```

## Sacar el vocabulario del txt

* generamos el vovabulario de todo el txt con set(), y lo limpiamos, 118 es el numero de palabras que existen

```python
vocab = set(text.lower().replace('\n', ' ').replace('.', '').replace(',', '').split())
len(vocab)
```

## Generamos el corpus a partir del txt

```python
corpus = text.split('\n')
```

```python
len(corpus)
```

* Funcion para obtener el vector TF de un documento

```python
def vector_TF(doc, vocab):
    vec =[]
    for word in vocab:
        cont = doc.lower().replace('.', '').replace(',', '').split().count(word)
        vec.append(cont)
    return np.array(vec)
```

```python
vector_TF(corpus[0],vocab)
```

* Aplicamos el vector TF a cada documento del corpus, y obtenemos la matriz TF del corpus

```python
def matrix_TF(corpus):
    corpus_vec = []
    for doc in corpus:
        doc_vec = vector_TF(doc)
        corpus_vec.append(doc_vec)
    return np.array(corpus_vec)
```

```python
matrix_TF(corpus)
```

* Generamos el vector IDF

```python
def vector_IDF(corpus):
    df_t = np.sum(matrix_TF(corpus), axis=0) # array
    return np.log(np.divide(len(corpus), df_t))
```

* DF_T es la suma de las columnas de la matrix TF

```python
Matrix_TF_IDF = matrix_TF(corpus) * vector_IDF(corpus)
```
